```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, rand, current_timestamp

# 1. Start the Spark Engine
spark = SparkSession.builder.appName("PortfolioBigData").getOrCreate()

# 2. Generate 1,000,000 rows of fake streaming log data
print("Spark Engine Online! Generating 1 Million rows...")

df = spark.range(0, 1000000) \
    .withColumn("user_id", (rand() * 10000).cast("int")) \
    .withColumn("movie_id", (rand() * 500).cast("int")) \
    .withColumn("watch_duration_minutes", (rand() * 180).cast("int")) \
    .withColumn("timestamp", current_timestamp())

# 3. Show the first 10 rows
df.show(10)
```

```python
# The Executives want to know: What are the Top 10 most watched movies today?
top_movies = df.groupBy("movie_id") \
    .sum("watch_duration_minutes") \
    .withColumnRenamed("sum(watch_duration_minutes)", "total_watch_time") \
    .orderBy(col("total_watch_time").desc()) \
    .limit(10)

top_movies.show()
```
